# Исследовательская тетрадка по прогнозу пополнений банкоматов

```python
import os
import pandas as pd

from matplotlib import pyplot as plt
from plotly import express as px

from bank_schedule import data as bsd
```

```python
RAW_DATA_FOLDER = os.path.join('..', 'data', 'raw')
INTERIM_DATA_FOLDER = os.path.join('..', 'data', 'interim')
```

```python
loader = bsd.Data(RAW_DATA_FOLDER)
```

## Читаем данные

```python
in_df = loader.get_money_in()
start_df = loader.get_money_start()
geo_df = loader.get_geo_TIDS()

display(in_df.head())
print(in_df.info())

display(start_df.head())
print(start_df.info())

display(geo_df.head())
print(geo_df.info())
```

```python
money_in_geo_mean = in_df.groupby('TID')['money_in'].mean().reset_index()
money_in_geo_median = in_df.groupby('TID')['money_in'].median().reset_index()
money_in_geo_sum = in_df.groupby('TID')['money_in'].sum().reset_index()
```

## Визуализация
